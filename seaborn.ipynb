```python
import matplotlib.pyplot as plt
import seaborn as sns

iris = sns.load_dataset("iris")
```

```python
# ヒストグラム
sns.distplot(iris.sepal_length, kde = True, rug=False)
```

```python
# 散布図
sns.jointplot('sepal_width', 'petal_length', data=iris)
```

```python
# 散布図（全数値変数）
sns.pairplot(iris, hue="species")
```

```python
# 相関
# seaborn.heatmap(pd.DataFrame(X_train_scaled).corr(), annot=True)

plt.figure(figsize=(12, 12))
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
sns.heatmap(
    iris[columns].corr(),
    annot=True,
    # square=True,
    # fmt='.2f',
    annot_kws={'size': 15},
    # cmap='Accent'
)
```

```python
# lifelinesのサンプルデータ
# 元は犯罪者が再度逮捕されるまでの時間？

# cox回帰をかけると最終的に回帰に影響が大なのは、age**, prio**, fin*
# 色は緑が再逮捕された、青がされなかった
# age: 年齢
# prio: 前科の数
# fin: 財政支援の有無
# => see also lifelines.ipynb

from lifelines.datasets import load_rossi
rossi_dataset = load_rossi()

print(rossi_dataset.head())
sns.pairplot(rossi_dataset, hue="arrest", vars=['fin', 'age', 'race', 'wexp', 'mar', 'paro', 'prio'])
```

```python
plt.figure(figsize=(12, 12))
columns = ['arrest', 'fin', 'age', 'race', 'wexp', 'mar', 'paro', 'prio']
sns.heatmap(
    rossi_dataset[columns].corr(),
    annot=True,
    # square=True,
    # fmt='.2f',
    annot_kws={'size': 15},
    # cmap='Accent'
)
```
